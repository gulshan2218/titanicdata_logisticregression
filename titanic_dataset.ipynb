```python
import seaborn as sns
import pandas as pd
```

```python
titanic=sns.load_dataset('titanic')
```

```python
print(titanic.head())
```

```python
print(titanic.groupby('alone')['survived'].mean())
```

```python
age=pd.cut(titanic['age'],[0,18,50,100])
age
```

```python
titanic.pivot_table('survived',['sex',age],'class')
```

```python
titanic.head()
```

```python
sex = pd.get_dummies(titanic['sex'],drop_first=True)

titanic.drop(['sex'],axis=1,inplace=True)
titanic.head()
```

```python
titanic=pd.concat([titanic,sex],axis=1)
```

```python
titanic.drop(['adult_male','who'],axis=1,inplace=True)
```

```python
titanic.drop(['deck'],axis=1,inplace=True)
```

```python
#titanic=pd.concat([titanic,sex],axis=1)
```

```python
titanic.head()
```

```python
titanic.drop(['embarked','embark_town'],axis=1,inplace=True)
```

```python
titanic[['class']].replace('Second',2,inplace=True)
titanic[['class']].replace('First',1,inplace=True)
titanic[['class']].replace('Third',3,inplace=True)
```

```python
titanic[['class']].info()
```

```python
titanic.drop(['alive','alone'],axis=1,inplace=True)

titanic
```

```python
titanic.dropna(inplace=True)
titanic
```

```python
df_x =titanic.drop(['survived'],axis=1)
df_x
```

```python
df_y =titanic[['survived']]
df_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
y_predict=log.predict(x_test)
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_predict))
```
